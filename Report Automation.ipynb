```python
df1 = pd.read_excel("Collated Login Tracker.xlsx", sheet_name= "Branch & Pifa")
df = pd.read_excel("Collated Login Tracker.xlsx", sheet_name= "Call Center")
df1.head()
```

```python
df.head()
```

```python
final_df['BCSM ADID'] = ''
final_df['BCSM Name'] = ''
final_df['RCSM Name'] = ''
final_df['ZCSM Name'] = ''
final_df['Agent Code'] = ''
final_df['Name'] = ''
final_df['Floor Managers '] = ''
final_df['VP/HEAD'] = ''
```

```python
My_dic = dict(zip(df1['Product Holding Code (Deal No, Proposal ID, Application ID, Folio No, Finserv No, Policy NO, Loan No)'], df1['BCSM ADID']))
My_dic1 = dict(zip(df1['Product Holding Code (Deal No, Proposal ID, Application ID, Folio No, Finserv No, Policy NO, Loan No)'], df1['BCSM Name']))
My_dic2 = dict(zip(df1['Product Holding Code (Deal No, Proposal ID, Application ID, Folio No, Finserv No, Policy NO, Loan No)'], df1['RCSM Name']))
My_dic3 = dict(zip(df1['Product Holding Code (Deal No, Proposal ID, Application ID, Folio No, Finserv No, Policy NO, Loan No)'], df1['ZCSM Name']))
My_dic4 = dict(zip(df['Application Id/Proposer number'], df['Agent Code']))
My_dic5 = dict(zip(df['Application Id/Proposer number'], df['Name']))
My_dic6 = dict(zip(df['Application Id/Proposer number'], df['Floor Managers ']))
My_dic7 = dict(zip(df['Application Id/Proposer number'], df['VP/HEAD']))
```

```python
final_df['BCSM ADID'] = final_df['Product Holding Code'].map(My_dic)
final_df['BCSM Name'] = final_df['Product Holding Code'].map(My_dic1)
final_df['RCSM Name'] = final_df['Product Holding Code'].map(My_dic2)
final_df['ZCSM Name'] = final_df['Product Holding Code'].map(My_dic3)
final_df['BCSM ADID'] = final_df['Secondary product Holding Code (Loan No, Policy No, TRXN_NO)'].map(My_dic).fillna(final_df['BCSM ADID'])
final_df['BCSM Name'] = final_df['Secondary product Holding Code (Loan No, Policy No, TRXN_NO)'].map(My_dic1).fillna(final_df['BCSM Name'])
final_df['RCSM Name'] = final_df['Secondary product Holding Code (Loan No, Policy No, TRXN_NO)'].map(My_dic2).fillna(final_df['RCSM Name'])
final_df['ZCSM Name'] = final_df['Secondary product Holding Code (Loan No, Policy No, TRXN_NO)'].map(My_dic3).fillna(final_df['ZCSM Name'])
```

```python
final_df['Agent Code'] = final_df['Product Holding Code'].map(My_dic4)
final_df['Name'] = final_df['Product Holding Code'].map(My_dic5)
final_df['Floor Managers '] = final_df['Product Holding Code'].map(My_dic6)
final_df['VP/HEAD'] = final_df['Product Holding Code'].map(My_dic7)
final_df['Agent Code'] = final_df['Secondary product Holding Code (Loan No, Policy No, TRXN_NO)'].map(My_dic4).fillna(final_df['Agent Code'])
final_df['Name'] = final_df['Secondary product Holding Code (Loan No, Policy No, TRXN_NO)'].map(My_dic5).fillna(final_df['Name'])
final_df['Floor Managers '] = final_df['Secondary product Holding Code (Loan No, Policy No, TRXN_NO)'].map(My_dic6).fillna(final_df['Floor Managers '])
final_df['VP/HEAD'] = final_df['Secondary product Holding Code (Loan No, Policy No, TRXN_NO)'].map(My_dic7).fillna(final_df['VP/HEAD'])
```
